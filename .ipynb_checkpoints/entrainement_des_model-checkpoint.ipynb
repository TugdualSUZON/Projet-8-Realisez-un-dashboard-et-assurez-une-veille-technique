```python
def create_custom_scorer(metric):
    return make_scorer(lambda y_true, y_pred: apply_transformation_and_evaluate(y_true, y_pred, metric))
```

```python
pipe = Pipeline(
    [
        ("scaling", RobustScaler()),
        ("regressor", svm.SVC())
    ]
                )
```

```python
parameters = {
    "regressor__kernel": ("poly",), 
    "regressor__C": [2,4,6,8,10]
}

gridsearch = GridSearchCV(pipe, parameters)
```
